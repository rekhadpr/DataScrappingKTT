```python
def get_table(soup):
    titles=[i.text for i in soup.select("a.title.fw500.ellipsis")]
    company=[i.text for i in soup.select("div.mt-7.companyInfo.subheading.lh16")]
    exp=[]
    sal=[]
    loc=[]
    for i in soup.find_all("ul",class_="mt-7"):
        lis=i.find_all("li")
        exp.append(lis[0].text)
        sal.append(lis[1].text)
        loc.append(lis[2].text)
    table=pd.DataFrame({'title':titles,'company':company,'exp':exp,'sal':sal,'loc':loc})
    return table
```

```python
get_table(soup)
```

```python
import time
Tables=[]
for i in range(1,3):
    url=stub.format(i)
    browser.get(url)
    time.sleep(0.5)
    html=browser.page_source
    soup=BeautifulSoup(html,'html.parser')
    Tables.append(get_table(soup))
```

```python
pd.concat(Tables,axis=0)
```

```python
def get_html_table(soup_table):
    headers=[i.text.strip() for i in soup_table.thead.find_all("th")]
    data=[]
    for row in soup_table.tbody.find_all("tr"):
        data.append([i.text.strip() for i in  row.find_all("td")])
    table=pd.DataFrame(data)
    table.columns=headers
    return table
```

```python
stub="https://en.wikipedia.org/wiki/List_of_terrorist_incidents_in_{}"
```

```python
stub.format(1970)
```

```python
import time
T=[]
for i in range(1970,1980,1):
    url=stub.format(i)
    browser.get(url)
    time.sleep(1)
    html=browser.page_source
    soup=BeautifulSoup(html,'html.parser')
    tables=soup.select("table.wikitable.sortable.jquery-tablesorter")
    for table in tables:
        t=get_html_table(table)
        t['year']=i
        T.append(t)
```

```python
len(T)
```

```python
pd.concat(T,axis=0)
```
